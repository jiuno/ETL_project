```python
import pandas as pd

import os

import datetime

import sqlite3
```

```python
DB_PATH = "../local_db/bronze.db"
```

```python
path = "../sources/crm/cust_info.csv"
source_file = os.path.basename(path)
df = pd.read_csv(path)


df["ingestion_date"] = datetime.datetime.now().isoformat()
df["source_file"] = source_file

display(df)
```

```python
with sqlite3.connect(DB_PATH) as conn:
    df.to_sql(source_file, conn, if_exists="append", index=True)
```
